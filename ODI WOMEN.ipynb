```python
from bs4 import BeautifulSoup
import requests
```

```python
women_team =[]
matches =[]
points = []
ratings = []

page = requests.get('https://www.icc-cricket.com/rankings/womens/team-rankings/odi')

soup = BeautifulSoup(page.content)

top1 = soup.find_all('span', class_="u-hide-phablet")

for i in top1:
    women_team.append(i.text.replace('\n',''))

top2 = soup.find('td', class_="rankings-block__banner--matches")

matches.append(top2.text.replace('\n','').strip())

top3 = soup.find('td', class_="rankings-block__banner--points")

points.append(top3.text.replace('\n','').strip())

top4 = soup.find('td', class_ = "rankings-block__banner--rating u-text-right")

ratings.append(top4.text.replace('\n','').strip())

rest_top = soup.find_all('tr', class_="table-body")

for i in rest_top:
    w2 = matches.append(i.find('td', class_="table-body__cell u-center-text").text.replace('\n','').strip())
    w3 = points.append(i.find('td', class_="table-body__cell u-center-text").find_next_sibling('td').text.replace(',',''))
    w4 = ratings.append(i.find('td', class_="table-body__cell u-text-right rating").text.replace('\n','').strip())
```

```python
import pandas as pd


women_cric = pd.DataFrame({})
women_cric['women_team'] = women_team[0:10]
women_cric['matches'] = matches[0:10]
women_cric['points'] = points[0:10]
women_cric['ratings'] = ratings[0:10]

women_cric
```

# Question 2

```python
w_player =[]
team = []
rating = []

page = requests.get('https://www.icc-cricket.com/rankings/womens/player-rankings/odi/batting')

soup = BeautifulSoup(page.content)

p1 = soup.find('div', class_="rankings-block__banner--name-large")
w_player.append(p1.text.replace('\n',''))

p2 = soup.find('div', class_="rankings-block__banner--nationality")
team.append(p2.text.replace('\n','').strip())

p3 = soup.find('div', class_="rankings-block__banner--rating")
rating.append(p3.text.replace('\n','').strip())

p4 = soup.find_all('tr', class_="table-body")

for i in p4:
    w1 = w_player.append(i.a.text.replace('\n',''))
    w2 = team.append(i.find('span', class_="table-body__logo-text").text.replace('\n','').strip())
    w3 = rating.append(i.find('td', class_="table-body__cell rating").text.replace('\n','').strip())
```

```python
women_batsmen = pd.DataFrame({})
women_batsmen['Player'] = w_player[0:10]
women_batsmen['team'] = team[0:10]
women_batsmen['rating'] = rating[0:10]

women_batsmen
```
