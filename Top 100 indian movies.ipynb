```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://www.imdb.com/list/ls079077479/')
```

```python
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
Movies=soup.find('h3',class_="lister-item-header")
Movies
```

```python
Movies.a.text.replace('\n','')
```

```python
Rating=soup.find('div', class_="ipl-rating-star small")
Rating
```

```python
Rating.text.replace('\n','')
```

```python
Year=soup.find('span', class_="lister-item-year text-muted unbold")
Year
```

```python
Year.text
```

```python
Name=[]
for i in soup.find_all('h3', class_="lister-item-header"):
    Name.append(i.a.text.replace('\n',''))
Name
```

```python
StarRating=[]
for i in soup.find_all('div', class_="ipl-rating-star small"):
    StarRating.append(i.text.replace('\n',''))
StarRating
```

```python
year=[]
for i in soup.find_all('span', class_="lister-item-year text-muted unbold"):
    year.append(i.text)
year
```

```python
import pandas as pd
import numpy as np
df=pd.DataFrame({'Movies':Name,'Rating':StarRating,'Year':year})
df.index = np.arange(1, len(df)+1)
df
```
